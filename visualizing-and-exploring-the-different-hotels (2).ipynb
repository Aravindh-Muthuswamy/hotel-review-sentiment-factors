```python
import matplotlib.pyplot as plt
def print_block_per(from_val, to_val):
    for i in range(from_val,to_val):
        print(df1['hotel_name'].unique()[i])
        rating = df[df['Hotel_name'].str.contains(df1['hotel_name'].unique()[i])]['Rating_Percentage_clean']
        print(rating.value_counts().sort_values())
        print(rating.unique())
        rating_series = pd.Series(rating.unique())
        rating_tot = rating.value_counts().sort_index()
        rating_tot.plot(kind="bar")
        plt.show()
```

```python
df['sentiment_clean'] = pd.to_numeric(df['Sentiment'], errors='coerce')
df['sentiment_clean'].value_counts()
```

```python
def print_sent_per(from_val, to_val):
    for i in range(from_val,to_val):
        df_vis = pd.DataFrame()
#         print(df1['hotel_name'].unique()[i])
        hotel = df1['hotel_name'].unique()[i]
#         print(hotel)
        rating = df[df['Hotel_name'].str.contains(df1['hotel_name'].unique()[i])]['sentiment_clean']
#         print(rating.value_counts().sort_index())
        rating_series = pd.Series(rating.unique())
        rating_tot = rating.value_counts().sort_index()
        seriesx = [1,2,3,4]
        df_vis['hotel'] = seriesx
        df_vis['rating_tot'] = rating_tot
#         print(df_vis)
        df_vis['rating_tot'].plot(use_index=True, figsize=(10,10))
    plt.show()
```
